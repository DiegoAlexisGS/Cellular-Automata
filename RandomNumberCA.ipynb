```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
archivo = open("/home/diego/sketchbook/examples/sketch_180510a/a.txt", "r") 
contenido = archivo.read()
contenido=list(contenido.split("\n"))
```

```python
nums=[]
for m in range(len(contenido)-1):
    o=int(contenido[m],2)
    nums.append(o)
```

```python
df=pd.DataFrame(nums)
df.plot(kind="hist")
plt.show()
```

```python
a=df.apply(pd.value_counts)
fig,ax=plt.subplots()
a[0]
df.drop_duplicates(inplace=True)
ax.scatter(df,a)
plt.show()
```
